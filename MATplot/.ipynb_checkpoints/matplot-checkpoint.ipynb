```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,np.pi*2,100)
siny = np.sin(x)
plt.plot(x,siny)
plt.title("sine wave")
plt.show()
```

# cos values 


```python
x= np.linspace(0,np.pi*2,100)
siny = np.sin(x)
cosy = np.cos(x)
plt.plot(x,siny,label="sin")
plt.plot(x,cosy,label="cos")
plt.title("sine wave and cos wave")
plt.legend()

plt.show()
```

# figure size(figsize= (x,y))

```python
plt.figure(figsize=(20,10)) # x and y

x= np.linspace(0,np.pi*4,100)
siny = np.sin(x)
cosy = np.cos(x)
plt.style.use('grayscale')  # style use 
plt.plot(x,siny,label="sin")
plt.plot(x,cosy,label="cos")
plt.title("sine wave and cos wave")
plt.legend()

plt.show()
```

# save figure to local machine

```python
plt.figure(figsize=(20,10)) # x and y

x= np.linspace(0,np.pi*4,100)
siny = np.sin(x)
cosy = np.cos(x)

fig  = plt.figure()   # function to save image


plt.style.use('classic')  # style use 
plt.plot(x,siny,label="sin")
plt.plot(x,cosy,label="cos")
plt.title("sine wave and cos wave")


fig.savefig("sine and cos.png") # save png


plt.legend()

plt.show()
```

# Adjust a plot (xlim,ylim)

```python
x = np.linspace(0,20,150)
siny =  np.sin(x)

plt.title("sine wave")
plt.xlim([0,10]) # from 0 to 10 on x axis
plt.ylim([-2,2])
plt.plot(x,siny)
plt.show()
```

# scatter plot
c = color
alpha ranges from 0 to 1 
                    0 is transparent  and 1 is dark
s = size of scatter points

```python
x = np.random.randint(1,1000,500)
y = np.random.randint(1,1000,500)
plt.scatter(x,y, c="red",alpha = 0.5 ,s =20)
plt.style.use("classic")
plt.title("scatter plot")

plt.show()
```

# multiple plots


```python
plt.figure(figsize=(15,8))
x = np.random.randint(1,1000,500)
y = np.random.randint(1,1000,500)
plt.scatter(x,y, c="red",alpha = 0.5 ,s =20)
plt.scatter(y+800,x+800)
plt.title("Multiple scatter plot")

plt.show()
```

```python
rng = np.random.RandomState(0)
x=np.random.randint(0,100,100)
y=np.random.randint(0,100,100)
colors = rng.rand(100)

plt.scatter(x,y, c=colors, s = 1000 ,edgecolors ="black" ,alpha = 0.6 ,cmap='plasma')
plt.colorbar()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# PIECHART

```python
bikes = ["ktm390","BMW GSA 1250R","rtr200","benali 300", "kawasaki-ninja 900"]
counts = [30     ,10              ,50      , 15        ,     12]
plt.figure(figsize=(7,7),facecolor="white")
plt.pie(counts , labels =bikes )
plt.show()
```


explode

```python
bikes = ["ktm390","BMW GSA 1250R","rtr200","benali 300", "kawasaki-ninja 900"]
counts = [30     ,10              ,50      , 15        ,     12]
explode =  [0,0.1,0,0,0]  # it explodes bmw by 10%
plt.figure(figsize=(7,7),facecolor="white" )
plt.pie(counts , labels =bikes ,explode = explode )
plt.show()
```

SHADOW


```python
bikes = ["ktm390","BMW GSA 1250R","rtr200","benali 300", "kawasaki-ninja 900"]
counts = [30     ,10              ,50      , 15        ,     12]
explode =  [0,0.1,0,0,0]  # it explodes bmw by 10%
plt.figure(figsize=(7,7),facecolor="white")

plt.pie(counts , labels =bikes ,explode = explode , shadow = True)
plt.show()
```

PERCENTAGE  code is autopct = %1.1f%%

```python
bikes = ["ktm390","BMW GSA 1250R","rtr200","benali 300", "kawasaki-ninja 900"]
counts = [30     ,10              ,50      , 15        ,     12]
explode =  [0,0.1,0,0,0]  # it explodes bmw by 10%
plt.figure(figsize=(7,7),facecolor="white")

plt.pie(counts , labels =bikes ,explode = explode , shadow = True , autopct = '%1.1%% )
plt.show()
```
